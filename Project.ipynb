```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
dataframe= pd.read_csv('/content/flights_data.csv')
print(dataframe.shape)
dataframe.head()
```

```python
sb.countplot(data=dataframe, x='Source') 
plt.xlabel("Source",fontsize=12)
plt.ylabel("Number of flights",fontsize=12);
```

```python
base_color=sb.color_palette()[0]
sb.countplot(data=dataframe, x='Source',  color= base_color)
plt.xticks(rotation=90,fontsize=12)
```

```python
base_color=sb.color_palette()[1]
gen_order=dataframe['Source'].value_counts().index
sb.countplot(data= dataframe, x='Source' , order=gen_order,  color=base_color)
```

```python
base_color=sb.color_palette()[2]
sb.countplot(data=dataframe, x='Airline', color=base_color)
plt.xticks(rotation=90)
plt.ylabel("Number of flights")
```

```python
base_color=sb.color_palette()[2]
sb.countplot(data=dataframe, y="Airline", color=base_color)
plt.xlabel("Number of flights")
```

```python
dataframe.isna().sum()
```

```python
null_count=dataframe.isna().sum()
base_color=sb.color_palette()[0]
sb.barplot(null_count.index.values,null_count,color=base_color)
plt.xticks(rotation=90)
plt.ylabel("Missing Values",fontsize=12);
```

```python
sorted_counts=dataframe['Destination'].value_counts()
plt.pie(sorted_counts,labels=sorted_counts.index,startangle=90,counterclock=False)
plt.axis('square')
plt.title('Flights Destination')
```

```python
sorted_counts=dataframe['Destination'].value_counts()
plt.pie(sorted_counts,labels=sorted_counts.index,startangle=90,counterclock=False,wedgeprops={'width':0.5})
plt.axis('square');
```

```python
plt.hist(data=[dataframe],x='Durations(minutes)')
```

```python
plt.hist(data=dataframe , x='Price', bins=20)
```

```python
Bins=np.arange(0,dataframe['Price'].max()+1,1200)
plt.hist(data=dataframe,x='Price',bins=Bins)
plt.show()
```

```python
bin_edges=np.arange(0,dataframe['Price'].max()+1,1200)
sb.distplot(dataframe['Price'],bins=bin_edges,kde=False,hist_kws={'alpha':1});
```
